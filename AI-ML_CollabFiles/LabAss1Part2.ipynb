```python
import os
import cv2
from google.colab.patches import cv2_imshow as imshow2
```

```python
os.listdir("/content/")
```

```python
#Question 1,2
dataset = []
for LcdTv_Images in os.listdir("/content/drive/MyDrive/Colab Notebooks/Lab Ass2/LcdTv_Images"):
  image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Lab Ass2/LcdTv_Images/" + LcdTv_Images)
  dataset.append(image)
```

```python
for x in range(0,25):
  print(x+1)
  imshow2(dataset[x])
```

```python
for x in range(0,25):
  print(x+1," ",dataset[x].shape)
```

```python
for x in range(0,25):
  print(dataset[x])
```

```python
#Question 3(a)
for x in range(0,25):
  print(x+1)
  imshow2(cv2.resize(dataset[x], (300, 350)))
```

```python
#Question 3(b)
for x in range(0,25):
  crop_img = dataset[x][50:150,100:250]
  print(x+1)
  imshow2(crop_img)
```

```python
#Question 4(a)
for x in range(0,25):
  gray_img = cv2.cvtColor(dataset[x], cv2.COLOR_RGB2GRAY)
  print(x+1)
  imshow2(gray_img)
  cv2.imwrite("/content/drive/MyDrive/Colab Notebooks/Lab Ass2/Dataset_1/grayimg_dataset{}.jpg".format(x),gray_img)
```

```python
#Question 4(b)
for x in range(0,25):
  binary_img = cv2.threshold(dataset[x], 150,255,cv2.THRESH_BINARY)[1]
  print(x+1)
  imshow2(binary_img)
  cv2.imwrite("/content/drive/MyDrive/Colab Notebooks/Lab Ass2/Dataset_1/binimg_dataset{}.jpg".format(x),binary_img)
```
