```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/speeddating.csv",na_values="?")
data
```

```python
data = data.drop(columns=['has_null','wave'])
data.head()
```

```python
X = data.drop(columns='match')
Y = data['match']
```

```python
all_col = X.columns
```

```python
for col in all_col:
  if(type(X[col][0])==str):
    if'['in list(X[col][0]):
      str1 = X[col][0].split('-')
      str1_0 = str1[0].split('[')
      str1_1 = str1[1].split(']')
      for i in range(8378):
        j = float(str1_0[1])
        k = float(str1_1[0])
        avg = (k+j)/2
        X[col][i] = int(avg)
```

```python
from sklearn.preprocessing import LabelEncoder
data_string = data.dtypes==object
data_cols = data.columns[data_string].tolist()
le = LabelEncoder()
data[data_cols] = data[data_cols].apply(lambda col: le.fit_transform(col))
data
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')
enc_df = pd.DataFrame(enc.fit_transform(data).toarray())
enc_df
```

```python
one_hot_encoded_data = pd.get_dummies(data, columns = ['race','race_o'])
one_hot_encoded_data
```

```python
race_scalar = np.array(data['race'])
normalized_arr = preprocessing.normalize([race_scalar])
print(normalized_arr)
```
