```python
# NOTE: DO NOT CHANGE THE VARIABLE NAME(S) IN THIS CELL
# Load the data
#QUESTION:1 and 2
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/student-mat.csv",delimiter =";",usecols=['age','address','famsize','reason','studytime','failures','schoolsup','famsup','paid','activities','higher','internet','romantic','freetime','goout','health','absences','G1','G2','G3'])
data
```

```python
data.shape
```

```python
df = data.iloc[:,:-1]
```

```python
# Handle categorical values
X = pd.get_dummies(df,columns=['address','famsize','reason','studytime','failures','schoolsup','famsup','paid','activities','higher','internet','romantic','freetime','goout','health','absences','G1','G2'],drop_first=True)
X
```

```python
y = data.iloc[:,-1]
y
```

```python
#QUESTION:3
X.columns
```

```python
for i in X.columns:
  X[i]=minmax_scale(X[i])
X
```

```python
#QUESTION:4
# Initialize the model
model =  LinearRegression()
```

```python
# Fit the model. Wait! We will complete this step for you ;)
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
#QUESTION:5
# Print mean squared error
mse = mean_squared_error(y,y_pred)

print(f"\nMSE: {mse}")
```
